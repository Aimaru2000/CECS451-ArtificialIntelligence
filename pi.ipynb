```python
# Sydney Kao
# 017615869
# CECS 451 - 01
# Professor Moon
# Assignment 7

import math
import random
import matplotlib.pyplot as plt

# n refers to 10^n
def estimate_pi(n):
    # x coordinates
    x = []    
    # y coordinates
    y = []
    
    # for points that are inside the quarter circle
    inside = []    
    # for points that are outside the quarter circle
    outside = []
    
    # counters for points inside and outside
    in_count = 0
    out_count = 0
    
    # generate random points
    for i in range(pow(10, n)):
        x_point = random.uniform(0, 1)
        y_point = random.uniform(0, 1)
        x.append(x_point)
        y.append(y_point)
        
    # check the distance of each point generated
    for i in range(pow(10, n)):
        distance = math.sqrt(pow(x[i], 2) + pow(y[i], 2))
        # if the distance is less than one, it is inside the circle
        if distance < 1:
            in_point = [x[i], y[i]]
            inside.append(in_point)
            in_count += 1
        # otherwise, it is outside the circle
        else:
            out_point = [x[i], y[i]]
            outside.append(out_point)
            out_count += 1
    
    # estimate the value of pi
    pi_value = 4 * ((in_count) / pow(10, n))
    
    # compute the error rates
    error = round(abs((pi_value - math.pi) / math.pi) * 100, 4)
    pi_value_output = round(pi_value, 6)
    
    print("n = 10 ^", n, " pi =", "{:.6f}".format(pi_value), " error =", error, "%")
    
    # return list of points inside the quarter circle, and list outside the quarter circle
    return inside, outside


# display the scatterplot
def displayValues(inside, outside):
    # generates tuples containing the points
    inside_points = zip(*inside)
    outside_points = zip(*outside)
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_aspect('equal', adjustable='box')
    
    plt.scatter(*inside_points, 0.5, marker='.', color='b')
    plt.scatter(*outside_points, 0.5, marker='.', color='r')
    plt.show()
    
    
# print the outputs for n = 10^3, 10^4, 10^5, 10^6
if __name__=="__main__":
    # print estimated value of pi
    points1 = estimate_pi(3)
    points2 = estimate_pi(4)
    points3 = estimate_pi(5)
    points4 = estimate_pi(6)
    
    # print the scatter plot for n = 10^4
    displayValues(points2[0], points2[1])
     
    

```
